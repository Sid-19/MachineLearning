## Matplot Lib Basics

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


#Basic Graphs

x = np.arange(1,4)
y = np.arange(11,14)
plt.plot(x,y,label="x+10")
plt.ylabel("y axis")
plt.xlabel("x axis")
plt.legend()
plt.title("First Graph")
plt.show()
```
